```python
import nltk
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
x = SentimentIntensityAnalyzer()
```

```python
df= pd.read_csv('customer_reviews.csv')
```

```python
df.head()
```

```python
def senti(text):
    xyz = x.polarity_scores(text)['compound']
    if xyz>=-0.5:
        return "Positive", xyz
    elif xyz<=-0.5:
        return "Negative",xyz
    else:
        return "Neutral",xyz
```

```python
df['sentiment'] =df['ReviewText'].apply(senti)
```

```python
for index, row in df.iterrows():
    print(f"Review:{row['ReviewText']}\n sentiment:{row['sentiment']}\n")
```
